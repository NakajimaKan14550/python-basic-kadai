```python
class Human:

    def __init__(self, name, age):
        self.name = name
        self.age = age

    def check_adult(self):
        if self.age >= 20:
            return f"{self.name}さんは成人です"
        else:
            return f"{self.name}は未成年です"


user1 = Human("kan", 42)
user2 = Human("fuyu",4)

print(user1.check_adult())
print(user2.check_adult())
```
