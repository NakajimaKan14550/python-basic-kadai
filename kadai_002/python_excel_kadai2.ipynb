```python
import pandas as pd

# 1. バラバラの伝票（データ）を準備する
data = {
    '日付': ['2023-05-17', '2023-05-18', '2023-05-19', '2023-05-20', '2023-05-21'],
    '社員名': ['山田', '佐藤', '鈴木', '田中', '高橋'],
    '売上': [100, 200, 150, 300, 250],
    '部門': ['メーカー', '代理店', 'メーカー', '商社', '代理店']
}

# 2. 作業員(pd.DataFrame)が台帳（df）に書き写す
df = pd.DataFrame(data)

df['平均売上'] = df['売上'].mean()

print(df['平均売上'])

# 3. 完成した台帳を確認する
#print(df)



def judge_rank(sale):
  if sale >= 250:
    return 'A'
  elif sale >=200:
    return 'B'
  else:
    return 'C'


df['業績ランク']=df['売上'].apply(judge_rank)

writer = pd.ExcelWriter('業績レポート.xlsx')
df.to_excel(writer, sheet_name='Sheet1', index=False)
writer.close()


print("Excelへの追加が完了しました！")
```
